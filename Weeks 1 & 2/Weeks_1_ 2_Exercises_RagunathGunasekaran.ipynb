## d. Data Wrangling with Python: Activity 4 - Activity 04 Design your own CSV parser

```python
from itertools import zip_longest
```

```python
def dict_from_csv_line(header, line):
    # Zip them
    zipped_line = zip_longest(header, line, fillvalue=None)
    # Use dict comprehension to generate the final dict
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict
```

```python
### For this Activity we are not gonna make smaller subtasks like the ones before.
### Write the whole activity's code bellow this line

# copied the file and pasted in my local

with open("C:/Users/ragun/Documents/GitHub/DSC540/Weeks 1 & 2/sales_record.csv", "r") as fd:
    firstLine = fd.readline()
    header = firstLine.replace("\n", "").split(",")
    for i, line in enumerate(fd):
        # Here we loop over the first 10 lines in order to not to make the output too big
        line = line.replace("\n", "").split(",")
        dict_output = dict_from_csv_line(header, line)
        print(dict_output)
        
        if i > 5:
            break
```
